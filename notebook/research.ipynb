## Loading the Third Data-frame Ratings.csv

```python
rating_df = pd.read_csv('BX-Book-Ratings.csv', sep = ";", on_bad_lines = 'skip', encoding = 'latin-1')
```

```python
rating_df.head()
```

```python
rating_df.shape
```

```python
rating_df.rename(columns = {"User-ID":"User_id",
                            "Book-Rating":"Book_rating"}, inplace = True)
```

```python
rating_df.head()
```

```python
# Printing the 3 dataframe shape

print(book_df.shape, user_df.shape, rating_df.shape, sep = '\n')
```

```python
# Let's store users who had atleast rated more than 200 books

x = rating_df['User_id'].value_counts() > 200
x
```

```python
x[x].shape
```

```python
# Index no. of users who had read more than 200 books

y = x[x].index
y
```

```python
# creating another df who had read more than 200 books

rating_df = rating_df[rating_df['User_id'].isin(y)]
rating_df.head(10)
```

```python
# Now joining rating_df with book_df

rating_with_book = rating_df.merge(book_df, on = 'ISBN')
rating_with_book.head()
```

```python
# Let's look how many no. of rating does every books got

num_of_rating = rating_with_book.groupby('Title')['Book_rating'].count().reset_index()
num_of_rating.head(10)
```

```python
# Changing Book_rating column name to no. of rating
num_of_rating.rename(columns = {"Book_rating":"no_of_rating"}, inplace = True)
```

```python
num_of_rating.head()
```

```python
# merging this with rating with book data frame

final_rating = rating_with_book.merge(num_of_rating, on = 'Title')
final_rating.head()
```

```python
# Let's take those books which got only 50 or above 50 ratings

final_rating = final_rating[final_rating['no_of_rating'] >= 50]
final_rating.head()
```

```python
final_rating.shape
```

```python
# dropping the duplicates

final_rating.drop_duplicates(inplace = True)
final_rating.shape
```

```python
final_rating.isna().sum()
```

```python
# Let's create a Pivot table

book_pivot = final_rating.pivot_table(columns = 'User_id', index = 'Title', values = 'Book_rating')
book_pivot
```

```python
# converting NaN to zero first will replace NaN

book_pivot.fillna(0, inplace = True)
book_pivot
```

```python
# Will store the pivot book in sparse matrix

from scipy.sparse import csr_matrix

book_sparse = csr_matrix(book_pivot)

book_sparse
```

```python
type(book_sparse)
```

```python
# Now will import Nearest Neighbor Algorithm

from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm = 'brute')

model.fit(book_sparse)
```

```python
book_pivot.iloc[560,:]
```

```python
# Let's try to Recommend some books on the basis of distance and suggestion

distance, suggestion = model.kneighbors(book_pivot.iloc[560,:].values.reshape(1, -1), n_neighbors = 8)
```

## Testing Model

```python
def recommended_book(book_name):
    book_id = int(np.where(book_pivot.index == book_name)[0][0])
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1, -1), n_neighbors = 8)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You Searched for: '{book_name}' \n")
                print("The Suggested books are: \n")
            else:
                print(j)
```
